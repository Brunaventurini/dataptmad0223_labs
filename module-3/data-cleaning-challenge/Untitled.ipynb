```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('iris-data.csv')

# Display the first few rows of the dataset
df.head()

# Check for missing values
df.isnull().sum()

# Replace missing values with mean
df['sepal_length_cm'].fillna(df['sepal_length_cm'].mean(), inplace=True)
df['sepal_width_cm'].fillna(df['sepal_width_cm'].mean(), inplace=True)
df['petal_length_cm'].fillna(df['petal_length_cm'].mean(), inplace=True)
df['petal_width_cm'].fillna(df['petal_width_cm'].mean(), inplace=True)

# Remove outliers using domain knowledge
df = df[(df['sepal_length_cm'] >= 0) & (df['sepal_length_cm'] <= 10)]
df = df[(df['sepal_width_cm'] >= 0) & (df['sepal_width_cm'] <= 5)]
df = df[(df['petal_length_cm'] >= 0) & (df['petal_length_cm'] <= 7)]
df = df[(df['petal_width_cm'] >= 0) & (df['petal_width_cm'] <= 3)]

# Encode categorical variables
df['class'] = df['class'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2})

# Plot histograms
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='sepal_length_cm', kde=True)
plt.subplot(2, 2, 2)
sns.histplot(data=df, x='sepal_width_cm', kde=True)
plt.subplot(2, 2, 3)
sns.histplot(data=df, x='petal_length_cm', kde=True)
plt.subplot(2, 2, 4)
sns.histplot(data=df, x='petal_width_cm', kde=True)
plt.show()

# Plot boxplots
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='sepal_length_cm')
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='sepal_width_cm')
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='petal_length_cm')
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='petal_width_cm')
plt.show()

# Save the cleaned dataset to a CSV file
df.to_csv('cleaned_iris_dataset.csv', index=False)
```
