#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y1 = x**2
y2 = np.exp(x)

# Subplot 1 - Normal Scale
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'r', label='y=x^2', color='magenta')
ax1.plot(x, y2, 'b', label='y=e^x', color='blue')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('y=x^2 and y=e^x')
ax1.legend(loc='upper left')

# Subplot 2 - Logarithmic Scale
fig, ax2 = plt.subplots()
ax2.plot(x, y1, 'r', label='y=x^2', color='magenta')
ax2.plot(x, y2, 'b', label='y=e^x', color='blue')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('y=x^2 and y=e^x (logarithmic scale)')
ax2.set_xscale('linear')
ax2.set_yscale('log')
ax2.legend(loc='upper left')

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# convert the Date column to a datetime object
fitbit_data['Date'] = pd.to_datetime(fitbit_data['Date'], format='%Y-%m-%d')

# add a Month column to the DataFrame
fitbit_data['Month'] = fitbit_data['Date'].dt.month_name()

# add a Months_encoded column to the DataFrame
fitbit_data['Months_encoded'] = fitbit_data['Date'].dt.month

# add a Week_or_Weekend column to the DataFrame
fitbit_data['Week_or_Weekend'] = fitbit_data['Date'].dt.dayofweek

# map the Week_or_Weekend column to a new column called Day
fitbit_data['Day'] = fitbit_data['Week_or_Weekend'].map({0: 'Weekday', 1: 'Weekday', 2: 'Weekday', 3: 'Weekday', 4: 'Weekday', 5: 'Weekend', 6: 'Weekend'})

# create a new DataFrame with the median steps by month
monthly_median_steps = fitbit_data.groupby('Month')['Steps'].median().reset_index()

# create a new DataFrame with the median steps by day
daily_median_steps = fitbit_data.groupby('Day')['Steps'].median().reset_index()

# create the figure and axes
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# plot the median steps by month
plt.sca(axs[0])
plt.plot(monthly_median_steps['Month'], monthly_median_steps['Steps'], marker='o')
plt.title('Median Steps by Month')
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.xticks(rotation=45)

# plot the median steps by day
plt.sca(axs[1])
plt.bar(daily_median_steps['Day'], daily_median_steps['Steps'])
plt.title('Median Steps by Day')
plt.xlabel('Day')
plt.ylabel('Median Steps')

# show the plots
plt.show()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
plt.sca(axs[0])
plt.plot(monthly_median_steps['Month'], monthly_median_steps['Steps'], marker='o')
plt.title('Median Steps by Month')
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.xticks(rotation=45)


# MEDIAN STEPS BY WORK_OR_WEEKEND
plt.sca(axs[1])
plt.bar(daily_median_steps['Day'], daily_median_steps['Steps'])
plt.title('Median Steps by Day')
plt.xlabel('Day')
plt.ylabel('Median Steps')
```
